```python
import tensorflow as tf
from tensorflow.keras.datasets import cifar10
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
```

```python
#Loading CIFAR-10 dataset
(x_train,y_train),(x_test,y_test) = cifar10.load_data()
```

```python
#Normalize pixel values to [0,1]
# CIFAR-10 mean and std values
cifar10_mean = [0.4914, 0.4822, 0.4465]
cifar10_std = [0.2023, 0.1994, 0.2010]

# Normalize
x_train = (x_train / 255.0 - cifar10_mean) / cifar10_std
x_test = (x_test / 255.0 - cifar10_mean) / cifar10_std
```

```python
y_train
```

```python
#one-hot encoding the target attribute
y_train = to_categorical(y_train,10)
y_test = to_categorical(y_test,10)
```

```python
#printing training data and testing data shapes
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python


model = Sequential([
    Conv2D(32, (3, 3), activation='relu', padding='same', input_shape=(32, 32, 3)),
    MaxPooling2D(pool_size=(2, 2)),
    Dropout(0.25),

    
   #2nd layer 
    Conv2D(64, (3, 3), activation='relu', padding='same'),
    MaxPooling2D(pool_size=(2, 2)),
    Dropout(0.25),

#3rd layer
    Conv2D(128, (3, 3), activation='relu', padding='same'),
    MaxPooling2D(pool_size=(2, 2)),
    Dropout(0.25),

    Flatten(),
    Dense(128, activation='relu'),
    
    Dropout(0.5),
    Dense(10, activation='softmax')
])
```

```python
#Model Compile
model.compile(optimizer='adam', loss = 'categorical_crossentropy',metrics= ['accuracy'])
```

```python
#model summary
model.summary()
```

```python
#Train the model
his = model.fit(
    x_train,y_train,
    validation_split=0.2,
    epochs =10,
    batch_size =64,
    verbose =1
)
```

```python
#Evaluate the baseline model
loss,accuracy = model.evaluate(x_test,y_test,verbose=0)
print(f"Test accuracy of Base line model:{accuracy:.4f}")
```

```python
import matplotlib.pyplot as plt

# Plot training & validation accuracy
plt.figure(figsize=(14, 5))

plt.subplot(1, 2, 1)
plt.plot(his.history['accuracy'], label='Training Accuracy', marker='o')
plt.plot(his.history['val_accuracy'], label='Validation Accuracy', marker='o')
plt.title('Model Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend(loc='lower right')
plt.grid(True)
```

```python
# Plot training & validation loss
plt.subplot(1, 2, 2)
plt.plot(his.history['loss'], label='Training Loss', marker='o')
plt.plot(his.history['val_loss'], label='Validation Loss', marker='o')
plt.title('Model Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend(loc='upper right')
plt.grid(True)

plt.tight_layout()
plt.show()
```
